```python
import numpy as np
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import rdDepictor
class Retrosynthesis:
    """A simple example class"""
    def __init__(self, mol):
        self.mols = []
        self.mols.append(mol)
        self.comments = []
        self.comments.append("__init__")
        self.heads = []
        self.atom_block = []
        self.bond_block = []
        self.tails = []
        self.n_atoms = 0
        self.n_bonds = 0
        self.matrix = np.zeros([mol.GetNumAtoms(), mol.GetNumAtoms()])
        self.visible_bonds = []
        self.atom_symbols = []
        self.InputMolBlock(Chem.MolToMolBlock(mol))
    
    def InputMolBlock(self, molblock):
        for i, line in enumerate(molblock.split("\n")):
            if i < 4:
                #print("heads", line)
                self.heads.append(line)
                if i == 3:
                    n_atoms, n_bonds = [int(n) for n in line.split()[:2]]
            elif i < (4 + n_atoms):
                #print("atoms", line)
                self.atom_block.append(line)
                a = line.split()
                self.atom_symbols.append(a[3])
            elif i < (4 + n_atoms + n_bonds):
                self.bond_block.append(line)
                self.visible_bonds.append(True)
                n1, n2 = line.split()[:2]
                n1, n2 = int(n1) - 1, int(n2) - 1
                self.matrix[n1][n2] = int(len(self.bond_block))
                self.matrix[n2][n1] = int(len(self.bond_block))
            else:
                #print("tail", line)
                #print(line)
                if 'CHG' in line.split(): continue
                self.tails.append(line)
        #self.visible_atoms = [True for n in range(self.n_atoms)]
        self.n_atoms = n_atoms
        self.n_bonds = n_bonds
        
    def GetMolfile(self):
        #print(self.n_bonds, self.visible_bonds)
        out_str = ''
        #out_str += "\n".join(self.heads) + "\n"
        for i, line in enumerate(self.heads):
            if i == 3:
                out_str += "%3d%3d  0  0  0  0  0  0  0  0999 V2000\n" % (self.n_atoms, self.n_bonds)
            else:
                out_str += line + "\n"
        out_str += "\n".join(self.atom_block) + "\n"
        for line, visible in zip(self.bond_block, self.visible_bonds):
            if visible: out_str += line + "\n"
        out_str += "\n".join(self.tails) + "\n"
        return out_str
    
    def Reaction(self, reactiontype):
        if reactiontype == 0: # 脱水素的結合生成反応 A-B <= A + B
            pass
            
    
    def PrintMatrix(self):
        print(self.matrix)
        
    def DeleteBond(self, i): # 脱水素的結合生成反応 A-B <= A + B
        self.visible_bonds[int(i)] = False
        #print("delete_bond")
        self.n_bonds = sum([1 for n in self.visible_bonds if n == True])
        #print(self.n_bonds)
    
    def DeleteBonds(self, n):
        for idx, bond in enumerate(self.bond_block):
            new_molecule = Retrosynthesis(self.mol)
            b1 = line.split()
            
        
    def RestoreBond(self, i): # 水素化反応・脱水素化反応
        self.visible_bonds[i] = True
        self.n_bonds = sum([1 for n in self.visible_bonds if n == True])
        
    def AddBond(self, i, j, k):
        i = int(i)
        j = int(j)
        if self.matrix[i - 1][j - 1] == 0:
            #print(self.bond_block)
            #print(len(self.bond_block))
            self.bond_block.append("%3d%3d%3d  0" % (i, j, k))
            self.visible_bonds.append(True)
            self.matrix[i - 1][j - 1] = int(len(self.bond_block))
            self.matrix[j - 1][i - 1] = int(len(self.bond_block))
            self.n_bonds = sum([1 for n in self.visible_bonds if n == True])
        else:
            self.bond_block[int(self.matrix[i - 1][j - 1] - 1)] = "%3d%3d%3d  0" % (i, j, k)
            
    def FindSeq(self, length):
        #seqs = []
        seen = []
        for i in range(len(self.atom_block)):
            queue = []
            for j in range(len(self.atom_block)):
                if self.matrix[i][j] == 0: continue
                queue.append([i, j])
            while len(queue) > 0:
                #print(queue)
                seq = queue.pop()
                if len(seq) == length:
                    if seq in seen: 
                        continue
                    else:
                        seen.append(seq)
                        yield seq
                    continue
                k = seq[-1]
                for h in range(len(self.atom_block)):
                    if h in seq: continue
                    if self.matrix[k][h] == 0: continue
                    queue.append(seq + [h])

    def FindUniqSeq(self, length):
        seen = []
        for seq in self.FindSeq(length):
            reversed_seq = seq[::-1]
            #print(seq, reversed_seq)
            if seq in seen: continue
            if reversed_seq in seen: continue
            seen.append(seq)
            seen.append(reversed_seq)
            yield seq
            
    def FindRing(self, length):
        #seqs = []
        seen = []
        for i in range(len(self.atom_block)):
            queue = []
            for j in range(len(self.atom_block)):
                if self.matrix[i][j] == 0: continue
                queue.append([i, j])
            while len(queue) > 0:
                #print(queue)
                seq = queue.pop()
                if len(seq) == length + 1:
                    continue
                k = seq[-1]
                for h in range(len(self.atom_block)):
                    if self.matrix[k][h] == 0: continue
                    if h in seq: 
                        if len(seq) == length and seq[0] == h:
                            #print(seq, h)
                            if seq in seen:
                                continue
                            else:
                                seen.append(seq)
                                yield(seq)
                        continue
                    queue.append(seq + [h])

    def FindUniqRing(self, length):
        seen = []
        for seq in self.FindRing(length):
            reversed_seq = seq[::-1]
            #print(seq, reversed_seq)
            if seq in seen: continue
            if reversed_seq in seen: continue
            seen.append(seq)
            seen.append(reversed_seq)
            yield seq
        
    def UnsaturateTransfer2(self, n): # 不飽和結合転位 A-B=C <= A=B-C
        idx = 0
        for seq in self.FindSeq(3):
            b1 = self.bond_block[int(self.matrix[seq[0]][seq[1]] - 1)].split()
            b2 = self.bond_block[int(self.matrix[seq[1]][seq[2]] - 1)].split()
            if (b1[2] != '1') or (b2[2] != '2'): continue
            if idx == n:
                self.AddBond(seq[0] + 1, seq[1] + 1, 2)
                self.AddBond(seq[1] + 1, seq[2] + 1, 1)
                return True
            idx += 1
        return False
    
    def FindSingleDouble4(self):
        for seq in self.FindSeq(5):
            bond1 = int(self.matrix[seq[0]][seq[1]] - 1)
            bond2 = int(self.matrix[seq[1]][seq[2]] - 1)
            bond3 = int(self.matrix[seq[2]][seq[3]] - 1)
            bond4 = int(self.matrix[seq[3]][seq[4]] - 1)
            b1 = self.bond_block[bond1].split()
            b2 = self.bond_block[bond2].split()
            b3 = self.bond_block[bond3].split()
            b4 = self.bond_block[bond4].split()
            if b1[2] == '1' and b2[2] == '2' and b3[2] == '1' and b4[2] == '2':
                yield (seq)
    
    def UnsaturateTransfer4(self, n): # 不飽和結合転位 A-B=C-D=E <= A=B-C=D-E
        idx = 0
        for seq in self.FindSingleDouble4():
            if idx == n:
                bond1 = int(self.matrix[seq[0]][seq[1]] - 1)
                bond2 = int(self.matrix[seq[1]][seq[2]] - 1)
                bond3 = int(self.matrix[seq[2]][seq[3]] - 1)
                bond4 = int(self.matrix[seq[3]][seq[4]] - 1)
                b1 = self.bond_block[bond1].split()
                b2 = self.bond_block[bond2].split()
                b3 = self.bond_block[bond3].split()
                b4 = self.bond_block[bond4].split()
                self.AddBond(int(b1[0]), int(b1[1]), 2)
                self.AddBond(int(b2[0]), int(b2[1]), 1)
                self.AddBond(int(b3[0]), int(b3[1]), 2)
                self.AddBond(int(b4[0]), int(b4[1]), 1)
                return True
            idx += 1
        return False
    
    def FindSingleDouble6(self):
        for seq in self.FindSeq(7):
            bond1 = int(self.matrix[seq[0]][seq[1]] - 1)
            bond2 = int(self.matrix[seq[1]][seq[2]] - 1)
            bond3 = int(self.matrix[seq[2]][seq[3]] - 1)
            bond4 = int(self.matrix[seq[3]][seq[4]] - 1)
            bond5 = int(self.matrix[seq[4]][seq[5]] - 1)
            bond6 = int(self.matrix[seq[5]][seq[6]] - 1)
            b1 = self.bond_block[bond1].split()
            b2 = self.bond_block[bond2].split()
            b3 = self.bond_block[bond3].split()
            b4 = self.bond_block[bond4].split()
            b5 = self.bond_block[bond5].split()
            b6 = self.bond_block[bond6].split()
            if b1[2] == '1' and b2[2] == '2' and b3[2] == '1' and b4[2] == '2' and b5[2] == '1' and b6[2] == '2':
                yield(seq)
    
    def UnsaturateTransfer6(self, n): # 不飽和結合転位 A-B=C-D=E-F=G <= A=B-C=D-E=F-G
        idx = 0
        for seq in self.FindSingleDouble6():
            if idx == n:
                bond1 = int(self.matrix[seq[0]][seq[1]] - 1)
                bond2 = int(self.matrix[seq[1]][seq[2]] - 1)
                bond3 = int(self.matrix[seq[2]][seq[3]] - 1)
                bond4 = int(self.matrix[seq[3]][seq[4]] - 1)
                bond5 = int(self.matrix[seq[4]][seq[5]] - 1)
                bond6 = int(self.matrix[seq[5]][seq[6]] - 1)
                b1 = self.bond_block[bond1].split()
                b2 = self.bond_block[bond2].split()
                b3 = self.bond_block[bond3].split()
                b4 = self.bond_block[bond4].split()
                b5 = self.bond_block[bond5].split()
                b6 = self.bond_block[bond6].split()
                self.AddBond(int(b1[0]), int(b1[1]), 2)
                self.AddBond(int(b2[0]), int(b2[1]), 1)
                self.AddBond(int(b3[0]), int(b3[1]), 2)
                self.AddBond(int(b4[0]), int(b4[1]), 1)
                self.AddBond(int(b5[0]), int(b5[1]), 2)
                self.AddBond(int(b6[0]), int(b6[1]), 1)
                return True
            idx += 1
        return False
    
    def AromaticRingRotate6(self, n):
        idx = 0
        for ring in self.FindUniqRing(6):
            b1 = self.bond_block[int(self.matrix[ring[0]][ring[1]]) - 1].split()
            b2 = self.bond_block[int(self.matrix[ring[1]][ring[2]]) - 1].split()
            b3 = self.bond_block[int(self.matrix[ring[2]][ring[3]]) - 1].split()
            b4 = self.bond_block[int(self.matrix[ring[3]][ring[4]]) - 1].split()
            b5 = self.bond_block[int(self.matrix[ring[4]][ring[5]]) - 1].split()
            b6 = self.bond_block[int(self.matrix[ring[5]][ring[0]]) - 1].split()
            if b1[2] == '1' and b2[2] == '2' and b3[2] == '1' and b4[2] == '2' and b5[2] == '1' and b6[2] == '2':
                if idx == n:
                    print(ring)
                    print(b1, ring[0], ring[1])
                    print(b2, ring[1], ring[2])
                    print(b3, ring[2], ring[3])
                    print(b4, ring[3], ring[4])
                    print(b5, ring[4], ring[5])
                    print(b6, ring[5], ring[0])
                    self.AddBond(int(b1[0]), int(b1[1]), 2)
                    self.AddBond(int(b2[0]), int(b2[1]), 1)
                    self.AddBond(int(b3[0]), int(b3[1]), 2)
                    self.AddBond(int(b4[0]), int(b4[1]), 1)
                    self.AddBond(int(b5[0]), int(b5[1]), 2)
                    self.AddBond(int(b6[0]), int(b6[1]), 1)
                    return True
                idx += 1
            elif b1[2] == '2' and b2[2] == '1' and b3[2] == '2' and b4[2] == '1' and b5[2] == '2' and b6[2] == '1':
                if idx == n:
                    print(ring)
                    print(b1, ring[0], ring[1])
                    print(b2, ring[1], ring[2])
                    print(b3, ring[2], ring[3])
                    print(b4, ring[3], ring[4])
                    print(b5, ring[4], ring[5])
                    print(b6, ring[5], ring[0])
                    self.AddBond(int(b1[0]), int(b1[1]), 1)
                    self.AddBond(int(b2[0]), int(b2[1]), 2)
                    self.AddBond(int(b3[0]), int(b3[1]), 1)
                    self.AddBond(int(b4[0]), int(b4[1]), 2)
                    self.AddBond(int(b5[0]), int(b5[1]), 1)
                    self.AddBond(int(b6[0]), int(b6[1]), 2)
                    return True
                idx += 1
        return False
    
    def AdditionReaction(self, n): # 付加反応 A-B-C <= A=B + C
        idx = 0
        for seq in self.FindSingleSingle2():
            if idx == n:
                atm1, atm2, atm3 = seq
                bond1 = int(self.matrix[seq[0]][seq[1]] - 1)
                bond2 = int(self.matrix[seq[1]][seq[2]] - 1)
                self.AddBond(atm1 + 1, atm2 + 1, 2)
                self.DeleteBond(bond2)
                return True
            idx += 1
        return False
    
    def EliminationReaction(self, n): # 脱離反応 A=B + C <= A-B-C
        idx = 0
        for line in self.bond_block:
            b1 = line.split()
            if b1[2] != '2': continue
            for i, atm in enumerate(self.atom_block):
                if i == int(b1[0]) - 1: continue
                if i == int(b1[1]) - 1: continue
                if self.matrix[int(b1[0]) - 1][i] == 0:
                    if idx == n:
                        print("hoge")
                        self.AddBond(int(b1[0]), i + 1, 1)
                        self.AddBond(int(b1[0]), int(b1[1]), 1)
                        print(b1, i + 1, int(b1[0]), int(b1[1]) )
                        return True
                    idx += 1
                if self.matrix[int(b1[1]) - 1][i] == 0:
                    if idx == n:
                        self.AddBond(int(b1[1]), i + 1, 1)
                        self.AddBond(int(b1[0]), int(b1[1]), 1)
                        print(b1, i + 1, int(b1[0]), int(b1[1]) )
                        return True
                    idx += 1
        return False
    
    def RearrangementReaction(self, n): # 転位反応 A-B + C <= A + B-C
        idx = 0
        for bond1idx, line in enumerate(self.bond_block):
            b1 = line.split()
            if b1[2] != '1': continue
            for i, atm in enumerate(self.atom_block):
                if i == int(b1[0]) - 1: continue
                if i == int(b1[1]) - 1: continue
                if self.matrix[int(b1[0]) - 1][i] == 0:
                    if idx == n:
                        print("hoge")
                        self.AddBond(int(b1[0]), i + 1, 1)
                        self.DeleteBond(bond1idx)
                        print(b1, i + 1, int(b1[0]), int(b1[1]) )
                        return True
                    idx += 1
                if self.matrix[int(b1[1]) - 1][i] == 0:
                    if idx == n:
                        self.AddBond(int(b1[1]), i + 1, 1)
                        self.DeleteBond(bond1idx)
                        print(b1, i + 1, int(b1[0]), int(b1[1]) )
                        return True
                    idx += 1
        return False
    
    def DiOxidativeDigestion(self, n):
        bonds = []
        for bond1idx, bond in enumerate(self.bond_block):
            b1 = bond.split()
            if b1[2] == '2':
                a1 = self.atom_symbols[int(b1[0]) - 1]
                a2 = self.atom_symbols[int(b1[1]) - 1]
                if a1 == "O" and a2 != "O":
                    bonds.append([bond1idx, int(b1[0]), int(b1[1])])
                elif a1 != "O" and a2 == "O":
                    bonds.append([bond1idx, int(b1[1]), int(b1[0])])
        idx = 0
        for ary1 in bonds:
            for ary2 in bonds:
                if ary1[0] >= ary2[0]: continue
                if idx == n:
                    print (idx, ary1, ary2)
                    self.DeleteBond(ary1[0])
                    self.DeleteBond(ary2[0])
                    self.AddBond(ary1[1], ary2[1], 2)
                    self.AddBond(ary1[2], ary2[2], 2)
                    return True
                idx += 1
        return False
    
    def InsertOxygen(self, n): # 酸素挿入 A-O-B <= A-B + O
        seqs = []
        idx = 0
        for seq in self.FindUniqSeq(3):
            if self.atom_symbols[seq[1]] != 'O': continue
            if idx == n:
                bond1 = self.matrix[seq[0]][seq[1]] - 1
                bond2 = self.matrix[seq[1]][seq[2]] - 1
                self.DeleteBond(bond1)
                self.DeleteBond(bond2)
                self.AddBond(seq[0] + 1, seq[2] + 1, 1)
                return True
            idx += 1
        return False
    
    def DielsAlder(self, n):
        idx = 0
        for ring in self.FindRing(6):
            b1 = self.bond_block[int(self.matrix[ring[0]][ring[1]]) - 1].split()
            b2 = self.bond_block[int(self.matrix[ring[1]][ring[2]]) - 1].split()
            b3 = self.bond_block[int(self.matrix[ring[2]][ring[3]]) - 1].split()
            b4 = self.bond_block[int(self.matrix[ring[3]][ring[4]]) - 1].split()
            b5 = self.bond_block[int(self.matrix[ring[4]][ring[5]]) - 1].split()
            b6 = self.bond_block[int(self.matrix[ring[5]][ring[0]]) - 1].split()
            if b1[2] == '1' and b2[2] == '2' and b3[2] == '1' and b4[2] == '1' and b5[2] == '1' and b6[2] == '1':
                if idx == n:
                    self.AddBond(ring[0] + 1, ring[1] + 1, 2)
                    self.AddBond(ring[1] + 1, ring[2] + 1, 1)
                    self.AddBond(ring[2] + 1, ring[3] + 1, 2)
                    self.AddBond(ring[4] + 1, ring[5] + 1, 2)
                    self.DeleteBond(int(self.matrix[ring[3]][ring[4]]) - 1)
                    self.DeleteBond(int(self.matrix[ring[0]][ring[5]]) - 1)
                    return True
                #print(ring)
                idx += 1
        return False
    
    def RetroDielsAlder(self, n): # 逆ディールス・アルダー反応 C=C-C=C + C=C <= C1-C=C-C-C-C1
        idx = 0
        for seq in self.FindSeq(4):
            b1 = self.bond_block[int(self.matrix[seq[0]][seq[1]]) - 1].split()
            b2 = self.bond_block[int(self.matrix[seq[1]][seq[2]]) - 1].split()
            b3 = self.bond_block[int(self.matrix[seq[2]][seq[3]]) - 1].split()
            if (b1[2] != '2') or (b2[2] != '1') or (b3[2] != '2') : continue
            for bondidx, bond in enumerate(self.bond_block):
                b4 = bond.split()
                if b4[2] != '2': continue
                a1, a2 = int(b4[0]) - 1, int(b4[1]) - 1
                if (a1 in seq) or (a2 in seq): continue
                for bb in [[a1, a2], [a2, a1]]:
                    if idx == n:
                        self.AddBond(seq[0] + 1, seq[1] + 1, 1)
                        self.AddBond(seq[1] + 1, seq[2] + 1, 2)
                        self.AddBond(seq[2] + 1, seq[3] + 1, 1)
                        self.AddBond(seq[3] + 1, bb[0] + 1, 1)
                        self.AddBond(bb[0] + 1, bb[1] + 1, 1)
                        self.AddBond(bb[1] + 1, seq[0] + 1, 1)
                        return True
                    idx += 1
        return False
    
    def CopeRearrangement(self, n): # コープ転位・クライゼン転位 A=B-C-D-E=F <= C=B-A-F-E=D
        idx = 0
        for seq in self.FindSeq(6):
            b1 = self.bond_block[int(self.matrix[seq[0]][seq[1]]) - 1].split()
            b2 = self.bond_block[int(self.matrix[seq[1]][seq[2]]) - 1].split()
            b3 = self.bond_block[int(self.matrix[seq[2]][seq[3]]) - 1].split()
            b4 = self.bond_block[int(self.matrix[seq[3]][seq[4]]) - 1].split()
            b5 = self.bond_block[int(self.matrix[seq[4]][seq[5]]) - 1].split()
            if (b1[2] != '2') or (b2[2] != '1') or (b3[2] != '1') or (b4[2] != '1') or (b5[2] != '2'): continue
            if idx == n:
                print(seq)
                self.AddBond(seq[0] + 1, seq[1] + 1, 1) # AB
                self.AddBond(seq[1] + 1, seq[2] + 1, 2) # BC
                self.DeleteBond(int(self.matrix[seq[2]][seq[3]]) - 1) # CD
                self.AddBond(seq[3] + 1, seq[4] + 1, 2) # DE
                self.AddBond(seq[4] + 1, seq[5] + 1, 1) # EF
                self.AddBond(seq[0] + 1, seq[5] + 1, 1) # AF
                return True
            idx += 1 
        return False
   
    def Cycloaddition22(self, n): # [2,2]環化付加 A1-B-C-D1 <= A=B + C=D
        idx = 0
        for ring in self.FindRing(4):
            b1 = self.bond_block[int(self.matrix[ring[0]][ring[1]]) - 1].split()
            b2 = self.bond_block[int(self.matrix[ring[1]][ring[2]]) - 1].split()
            b3 = self.bond_block[int(self.matrix[ring[2]][ring[3]]) - 1].split()
            b4 = self.bond_block[int(self.matrix[ring[3]][ring[0]]) - 1].split()
            if (b1[2] != '1') or (b2[2] != '1') or (b3[2] != '1') or (b4[2] != '1'): continue
            if idx == n:
                self.AddBond(ring[0] + 1, ring[1] + 1, 2) # AB
                self.DeleteBond(int(self.matrix[ring[1]][ring[2]]) - 1) # BC
                self.AddBond(ring[2] + 1, ring[3] + 1, 2) # CD
                self.DeleteBond(int(self.matrix[ring[3]][ring[0]]) - 1) # DA
                return False
            idx += 1
        return False
            
```
